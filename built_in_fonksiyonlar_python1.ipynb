#PYTHON BUILT-IN Fonksiyonları ve Kısa Açıklaması

Daha fazla bilgi ve diğer fonksiyonların detayları için [Python resmi dökümantasyonunu](https://docs.python.org/3/library/functions.html) inceleyebilirsiniz.


### 1. `abs()`
Bir sayının mutlak değerini döndürür..
```python
print(abs(-5))  # Çıktı: 5
print(abs(3.7))  # Çıktı: 3.7
```

### 2. `all()`
Bir iterable'ın tüm elemanları True ise True döndürür..
```python
print(all([True, True, True]))  # Çıktı: True
print(all([True, False, True]))  # Çıktı: False
```

### 3. `any()`
Bir iterable'da en az bir eleman True ise True döndürür..
```python
print(any([False, False, True]))  # Çıktı: True
print(any([False, False, False]))  # Çıktı: False
```

### 4. `bin()`
Bir tam sayının ikilik tabanda gösterimini döndürür..
```python
print(bin(10))  # Çıktı: '0b1010'
```

### 5. `bool()`
Bir değeri boolean (True veya False) tipine dönüştürür.
```python
print(bool(0))  # Çıktı: False
print(bool(1))  # Çıktı: True
print(bool(""))  # Çıktı: False
```

### 6. `chr()`
Unicode tablosundaki sayısal değere karşılık gelen karakteri döndürür..
```python
print(chr(97))  # Çıktı: 'a'
```

### 7. `divmod()`
İki sayıyı bölüp, bölüm ve kalan olarak bir tuple döndürür..
```python
print(divmod(10, 3))  # Çıktı: (3, 1)
```

### 8. `enumerate()`
Bir iterable'ın indeks ve eleman çiftlerinden oluşan bir enumerate nesnesi döndürür..
```python
for index, value in enumerate(['a', 'b', 'c']):
    print(index, value)
# Çıktı:
# 0 a
# 1 b
# 2 c
```

### 9. `filter()`
Bir fonksiyonu uygulayarak, verilen iterable'daki elemanlardan True dönenleri filtreler.
```python
def is_even(n):
    return n % 2 == 0

print(list(filter(is_even, [1, 2, 3, 4, 5, 6])))  # Çıktı: [2, 4, 6]
```

### 10. `float()`
Bir değeri float (ondalık) tipine dönüştürür.
```python
print(float(5))  # Çıktı: 5.0
print(float("3.14"))  # Çıktı: 3.14
```

### 11. `hex()`
Bir tam sayının onaltılık (hexadecimal) gösterimini döndürür..
```python
print(hex(255))  # Çıktı: '0xff'
```

### 12. `int()`
Bir değeri tam sayı (integer) tipine dönüştürür.
```python
print(int(3.14))  # Çıktı: 3
print(int("10"))  # Çıktı: 10
```

### 13. `len()`
Bir objenin (string, list, tuple vb.) uzunluğunu döndürür..
```python
print(len("hello"))  # Çıktı: 5
print(len([1, 2, 3, 4]))  # Çıktı: 4
```

### 14. `list()`
Bir iterable'ı listeye dönüştürür.
```python
print(list("hello"))  # Çıktı: ['h', 'e', 'l', 'l', 'o']
print(list((1, 2, 3)))  # Çıktı: [1, 2, 3]
```

### 15. `max()`
Bir iterable'daki en büyük elemanı döndürür..
```python
print(max([1, 2, 3, 4, 5]))  # Çıktı: 5
print(max("hello"))  # Çıktı: 'o'
```

### 16. `min()`
Bir iterable'daki en küçük elemanı döndürür..
```python
print(min([1, 2, 3, 4, 5]))  # Çıktı: 1
print(min("hello"))  # Çıktı: 'e'
```

### 17. `map()`
Bir fonksiyonu, verilen iterable'ın her bir elemanına uygular ve sonuçları içeren bir map objesi döndürür..
```python
def square(n):
    return n * n

print(list(map(square, [1, 2, 3, 4])))  # Çıktı: [1, 4, 9, 16]
```

### 18. `ord()`
Bir karakterin Unicode tablosundaki sayısal değerini döndürür..
```python
print(ord('a'))  # Çıktı: 97
print(ord('A'))  # Çıktı: 65
```

### 19. `pow()`
Bir sayının, verilen üssü kadar kuvvetini alır.
```python
print(pow(2, 3))  # Çıktı: 8
print(pow(2, 3, 5))  # Çıktı: 3 (2^3 % 5)
```

### 20. `range()`
Belirtilen başlangıç(start), bitiş(stop) ve adımlarla(step) bir sayı dizisi döndürür..
```python
print(list(range(5)))  # Çıktı: [0, 1, 2, 3, 4]
print(list(range(1, 10, 2)))  # Çıktı: [1, 3, 5, 7, 9]
```

### 21. `round()`
Bir sayıyı belirtilen basamağa kadar yuvarlar.
```python
print(round(3.14159, 2))  # Çıktı: 3.14
print(round(2.718, 1))  # Çıktı: 2.7
```

### 22. `sorted()`
Bir iterable'ı sıralar ve yeni bir liste döndürür..
```python
print(sorted([3, 1, 4, 1, 5, 9]))  # Çıktı: [1, 1, 3, 4, 5, 9]
print(sorted("hello"))  # Çıktı: ['e', 'h', 'l', 'l', 'o']
```

### 23. `str()`
Bir değeri string (metin) tipine dönüştürür.
```python
print(str(123))  # Çıktı: '123'
print(str(3.14))  # Çıktı: '3.14'
```

### 24. `sum()`
Bir iterable'daki tüm elemanların toplamını döndürür..
```python
print(sum([1, 2, 3, 4]))  # Çıktı: 10
print(sum((10, 20, 30)))  # Çıktı: 60
```

### 25. `type()`
Bir objenin tipini döndürür..

### 26. `iter()`
Bir objeyi iterator'a dönüştürür.

### 28. `ascii()`

### 29. `callable()`
Bir objenin çağrılabilir (callable) olup olmadığını kontrol eder.

### 30. `classmethod()`
Bir metodu sınıf metodu olarak tanımlar.
```python
class MyClass:
    @classmethod
    def my_classmethod(cls):
        print("This is a class method.")

MyClass.my_classmethod()  # Çıktı: This is a class method.
```

```python
code = "print('Hello, world!')"
compiled_code = compile(code, '<string>', 'exec')
exec(compiled_code)  # Çıktı: Hello, world!
```

### 54. `property()`
Bir sınıf niteliği olarak çalışır.
```python
class MyClass:
    def __init__(self):
        self._x = None
    
    @property
    def x(self):
        return self._x
    
    @x.setter
    def x(self, value):
        self._x = value

obj = MyClass()
obj.x = 10
print(obj.x)  # Çıktı: 10
```

### 60. `staticmethod()

`

### 61. `super()`
Bir sınıfın ebeveyn veya kardeş sınıflarının bir yöntemini çağırır.
```python
class MyBaseClass:
    def hello(self):
        print("Hello from MyBaseClass")

class MySubClass(MyBaseClass):
    def hello(self):
        super().hello()
        print("Hello from MySubClass")

obj = MySubClass()
obj.hello()
# Çıktı:
# Hello from MyBaseClass
# Hello from MySubClass
```

### 62. `tuple()`
Bir tuple (demet) nesnesi oluşturur.
```python
print(tuple([1, 2, 3]))  # Çıktı: (1, 2, 3)
print(tuple("hello"))  # Çıktı: ('h', 'e', 'l', 'l', 'o')
```

### 63. `vars()`
Bir nesnenin __dict__ özniteliğini döndürür..
```python
class MyClass:
    def __init__(self):
        self.x = 10
        self.y = 20

obj = MyClass()
print(vars(obj))  # Çıktı: {'x': 10, 'y': 20}
```

### 64. `zip()`
Birden fazla iterable'ın elemanlarını bir araya getirir.
```python
a = [1, 2, 3]
b = ['a', 'b', 'c']
print(list(zip(a, b)))  # Çıktı: [(1, 'a'), (2, 'b'), (3, 'c')]
```

### 65. `__import__()`
Bir modül içe aktarma işlemi gerçekleştirir.
```python
math_module = __import__('math')
print(math_module.sqrt(16))  # Çıktı: 4.0
```
